```python

plt.figure(figsize=(10,6))  # Set the figure size
plt.hist(df['MonthlyIncome'], bins=30, color='skyblue')  # df is your DataFrame, 'MonthlyIncome' is the column
plt.xlabel('Monthly Income')  # Set the x-axis label
plt.ylabel('Frequency')  # Set the y-axis label
plt.title('Histogram of Monthly Income')  # Set the title of the plot
plt.show()
```

```python


# Create two subsets of data based on attrition
stayed = df[df['Attrition'] == 'No']['MonthlyIncome']
left = df[df['Attrition'] == 'Yes']['MonthlyIncome']

# Create histogram
plt.figure(figsize=(10,6))

plt.hist(stayed, bins=30, alpha=0.5, label='Stayed', color='blue')
plt.hist(left, bins=30, alpha=0.5, label='Left', color='red')

plt.xlabel('Monthly Income')
plt.ylabel('Frequency')
plt.title('Monthly Income vs Attrition')
plt.legend(loc='upper right')

plt.show()
```

```python
plt.hist(df['Age'], bins=20, color='blue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['DistanceFromHome'], bins=20, color='green', edgecolor='black')
plt.title('Distance From Home Distribution')
plt.xlabel('Distance From Home')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['YearsAtCompany'], bins=20, color='red', edgecolor='black')
plt.title('Years At Company Distribution')
plt.xlabel('Years At Company')
plt.ylabel('Frequency')
plt.show()
```

```python
for field in df['EducationField'].unique():
    plt.hist(df[df['EducationField']==field]['MonthlyIncome'], bins=20, label=field, alpha=0.5)
plt.title('Monthly Income by Education Field')
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
df['Department'].value_counts().plot(kind='bar', color='purple')
plt.title('Number of Employees in Each Department')
plt.xlabel('Department')
plt.ylabel('Frequency')
plt.show()
```
